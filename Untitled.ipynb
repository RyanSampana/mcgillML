```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.fromfile('train_x.bin', dtype='uint8')
x = x.reshape((100000,60,60))
import pandas as pd
df = pd.read_csv('./train_y.csv')
def display(i):
    img = x[i]
    plt.title('Example %d. Label: %d' % (i, df.Prediction[i]))
    plt.imshow(img.reshape((60,60)), cmap=plt.cm.gray_r) 
```
